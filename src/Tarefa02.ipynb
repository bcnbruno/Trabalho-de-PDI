```python
from PIL import Image as img
import numpy as np
import math
import random as rd
```

```python
corredor = img.open("corredor_claro.jpg")

#cria o mapa de pixels
def pixelMap(img):
    return img.load()

def sal_pimenta(image, porcentagem,pro):
    img = image.copy()
    pixels = pixelMap(img)
    num = int(img.size[0]*img.size[1]*porcentagem)
    p = True
    for i in range(num):    # for every col:
        a, b = rd.randint(0,(img.size[0]-1)),rd.randint(0,(img.size[1]-1))
        if p:
            pixels[a,b] = (255,255,255)
            p=False
        else:
            pixels[a,b] = (0,0,0)
            p=True
    return img

s_p = sal_pimenta(corredor, 0.01,50)
```
