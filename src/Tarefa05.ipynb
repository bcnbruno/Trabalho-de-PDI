```python
from PIL import Image as img
import numpy as np
import math
```

```python
papR = img.open("papagaio-R.jpg")
papG = img.open("papagaio-G.jpg")
papB = img.open("papagaio-B.jpg")
```

```python
def componentesRGB(im):
    ig = im.copy()
    largura, altura = ig.size
    n = ig.load()
    R = []
    G = []
    B = []
    for x in range(largura):
        for y in range(altura):
            r,g,b = n[x,y]            
            R.append(r)
            G.append(g)
            B.append(b)
    return R,G,B

def juntar(r, g, b, largura, altura):
    new = img.new('RGB', (largura, altura), 'white')
    n = new.load()
    for x in range(largura):
        for y in range(altura):            
            n[x, y] = r[x*(altura)+y], g[x*altura+y], b[x*altura+y]
    return new
```

```python
pr, g, b = componentesRGB(papR)
r, pg, b = componentesRGB(papG)
r, g, pb = componentesRGB(papB)
l, a = papR.size
l = int(l)
a = int(a)
papColorida = juntar(pr, pg, pb, l, a)
papColorida
```

```python
def mudaCor(im):
    ig = im.copy()
    largura, altura = ig.size
    n = ig.load()
    for x in range(largura):
        for y in range(altura):
            #r,g,b = im.getpixel((x, y))
            r,g,b = n[x,y]
            avg = int((r + g + b) / 3.0)
            n[x,y] = (avg, avg, avg)
    return ig
```

```python
nova = mudaCor(papColorida)
nova
```
